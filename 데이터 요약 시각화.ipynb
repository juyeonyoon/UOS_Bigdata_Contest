### 서울시 구별 장애인 거주 

```python
import requests
import json


# 서울 행정구역 json raw파일(githubcontent)
r = requests.get('https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json')
c = r.content
seoul_geo = json.loads(c)
```

```python
import folium

m = folium.Map(
    location=[37.559819, 126.963895],
    zoom_start=11, 
    tiles = 'cartodbpositron'
)

folium.GeoJson(
    seoul_geo,
    name='지역구',
    style_function = lambda x: {'fillColor':'blue' ,
    							'color':'blue',
                                'weight' :'2'}
).add_to(m)


for lat, long, num in zip(data_loc['Y'], data_loc['X'], data_loc["num"]):
    folium.CircleMarker([lat, long], color="black",fill ="black", radius = int(num)/1000, popup = num, fill_opacity=int(num)/30000, weight=2).add_to(m)

m
```

### 베이어프리 실외 문화생활 시각화

```python
outplace = pd.read_csv(path+"실외 문화생활 정보.csv")
outplace.head()
```

```python
outplace.columns
```

```python
outplace["CTPRVN_NM"].unique()
```

```python
out_Seoul = outplace.query("CTPRVN_NM == '서울특별시'")
out_Seoul.head()
```

```python
out_Seoul = out_Seoul[["LCLAS_NM", "MLSFC_NM", "FCLTY_NM", "CTPRVN_NM", "SIGNGU_NM", "FCLTY_ROAD_NM_ADDR", "FCLTY_LA", "FCLTY_LO"]]
out_Seoul.head()
```

```python
out_Seoul = out_Seoul.reset_index(drop=True)
```

```python
import folium

m = folium.Map(
    location=[37.559819, 126.963895],
    zoom_start=11, 
    tiles = 'cartodbpositron'
)

folium.GeoJson(
    seoul_geo,
    name='지역구',
    style_function = lambda x: {'fillColor':'blue' ,
    							'color':'blue',
                                'weight' :'2'}
).add_to(m)

for lat, long, name, loc in zip(out_Seoul['FCLTY_LA'], out_Seoul['FCLTY_LO'], out_Seoul["FCLTY_NM"], out_Seoul["FCLTY_ROAD_NM_ADDR"]):
    folium.Marker([lat, long], popup = name, tooltip = [name, loc]).add_to(m)

m
```

```python
from folium.plugins import MarkerCluster


m = folium.Map(
    location=[37.559819, 126.963895],
    zoom_start=12, 
    tiles='cartodbpositron'
)

folium.GeoJson(
    seoul_geo,
    name='지역구'
).add_to(m)

marker_cluster = MarkerCluster().add_to(m)

for lat, long in zip(out_Seoul['FCLTY_LA'], out_Seoul['FCLTY_LO']):
    folium.Marker([lat, long], icon = folium.Icon(color="green")).add_to(marker_cluster)

m
```

```python
out_Seoul["SIGNGU_NM"].unique()
```

```python
dic = {}
gu_li = []
num_li = []
for i in out_Seoul["SIGNGU_NM"].unique():
    gu_li.append(i)
    num_li.append(sum(out_Seoul["SIGNGU_NM"]==i))

dic["gu"] = gu_li
dic["num"] = num_li
```

```python
out_num = pd.DataFrame(dic)
out_num
```

```python
out_place_num = pd.merge(out_num, gu, how='inner', on="gu")
```
